```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore


def z_score_anomaly_detection(data, column, threshold=3):
    z_scores = zscore(data[column])
    anomalies = np.where(np.abs(z_scores) > threshold)
    return anomalies

# Метод IQR
def iqr_anomaly_detection(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    anomalies = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return anomalies

#  метод Z-оценки
z_anomalies = z_score_anomaly_detection(data, 'Amount')

#  IQR
iqr_anomalies = iqr_anomaly_detection(data, 'Amount')

# Визуализация результатов
sns.boxplot(x=data['Amount'])
plt.title("Boxplot для визуализации аномалий (с помощью IQR и Z-оценки)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Разделение данных на обучающие и тестовые наборы
X = data[['Amount']]  # В качестве признака используем только столбец 'Amount'
y = data['Class']  # Целевая переменная (класс)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели Isolation Forest
model = IsolationForest(contamination=0.01, random_state=42)
model.fit(X_train)

# Предсказания для тестового набора
y_pred = model.predict(X_test)

# Переводим метки аномалии в бинарную классификацию (1 - нормальные, 0 - аномалии)
y_pred_binary = np.where(y_pred == 1, 0, 1)  # 1 - нормальные, 0 - аномалии (из-за особенностей модели)

# Метрики
accuracy = accuracy_score(y_test, y_pred_binary)
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-score: {f1:.4f}')

# Матрица путаницы
conf_matrix = confusion_matrix(y_test, y_pred_binary)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Normal', 'Anomaly'], yticklabels=['Normal', 'Anomaly'])
plt.title("Матрица путаницы")
plt.xlabel("Предсказанный класс")
plt.ylabel("Истинный класс")
plt.show()

# Кривая ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_binary)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Кривая ROC')
plt.xlabel('Ложноположительная вероятность (FPR)')
plt.ylabel('Истинноположительная вероятность (TPR)')
plt.legend(loc='lower right')
plt.show()
```
